```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn import linear_model
```

```python
bbdata = pd.read_csv("D:\Datasets For ML & DS Practice\Baseball_League_Data.csv")
```

```python
bbdata.info()
```

```python
bbdata.describe()
```

```python
bbdata = bbdata.dropna()
```

```python
bbdata.info()
```

```python
#3 Independent Variables
```

```python
X = bbdata[['years','games','atbats']]
Y = bbdata['salary']
  
# with sklearn
regr = linear_model.LinearRegression()
regr.fit(X, Y)
 
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
 
```

```python
# with statsmodels
X = sm.add_constant(X) # adding a constant
  
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
  
print_model = model.summary()
print(print_model)
```
